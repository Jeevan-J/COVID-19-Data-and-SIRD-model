```python
import mysql.connector as connector
import numpy as np, pandas as pd, os

# SQL Database where we stored COVID19 Data
HOST=os.getenv('DATABASE_IP')
USER=os.getenv('DATABASE_USER')
PASS=os.getenv('DATABASE_PASS')
DATABASE=os.getenv('DATABASE_NAME')
```

```python
conn = connector.connect(host=HOST,user=USER,passwd=PASS,database=DATABASE)
SQL_Query = pd.read_sql_query("SELECT * FROM daily_totalcases_india",conn)
data = pd.DataFrame(SQL_Query)
```

```python
data
```

```python
data['day_of_infection'] = data.index
data
```

```python
data['a'] = 0.0
data['b'] = 0.0
data['c'] = 0.0
data['growth_factor'] = 0
data['growth_factor_analysis'] = 0
data['total_population'] = 300000
data['susceptible'] = data['total_population'] - data['total_confirmed_cases']
data
```

```python
data['active_cases'] = data['total_confirmed_cases']-data['total_recovered_cases']-data['total_deaths']
```

```python
for index in data.index:
    if index == 0:
        pass
    else:
        #print(data['deaths'][index]/data['total_confirmed_cases'][index])
        data['growth_factor'][index] = float(data['total_confirmed_cases'][index])/data['total_confirmed_cases'][index-1]
        if index > 7:
            temp = []
            for jj in range(7):
                if data['growth_factor'][index-jj] >= 1:
                    temp.append(1)
                if data['growth_factor'][index-jj] < 1:
                    temp.append(-1)
            if sum(temp) >= 0:
                data['growth_factor_analysis'][index] = 'Spreading/Outbreak'
            else:
                data['growth_factor_analysis'][index] = 'Stopping'
        data['a'][index] = ((data['susceptible'][index-1]-data['susceptible'][index])*data['total_population'][index])/(data['susceptible'][index]*data['active_cases'][index])
        data['b'][index] = (data['recovered_cases'][index])/(data['active_cases'][index])
        data['c'][index] = (data['deaths'][index])/(data['active_cases'][index])
data
```

```python
data['a1'] = 0.0
data['b1'] = 0.0
data['c1'] = 0.0
for index in data.index:
    if index == 0:
        pass
    else:
        data['a1'][index] = ((data['susceptible'][0]-data['susceptible'][index])*data['total_population'][index])/(data['susceptible'][index]*data['active_cases'][index]*data['day_of_infection'][index])
        data['b1'][index] = (data['total_recovered_cases'][index]-data['total_recovered_cases'][0])/(data['active_cases'][index]*data['day_of_infection'][index])
        data['c1'][index] = (data['total_deaths'][index]-data['total_deaths'][0])/(data['active_cases'][index]*data['day_of_infection'][index])
data
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Total population, N.
N = (1377492805 * 0.0002)
N = 300000
print("N = ",N)
# Initial number of infected and recovered individuals, I0 and R0.
I0, R0, D0 = 1, 0, 0
# Everyone else, S0, is susceptible to infection initially.
S0 = N - I0 - R0 - D0
# Contact rate, beta, and mean recovery rate, gamma, (in 1/days).
beta, gamma,alpha = 0.25232, 0.1266, 0.00276
# A grid of time points (in days)
t = np.linspace(0, 200, 200)

# The SIR model differential equations.
def deriv(y, t, N, beta, gamma, alpha):
    S, I, R, D = y
    dSdt = -beta * S * I / N
    dIdt = beta * S * I / N - gamma * I - alpha * I
    dRdt = gamma * I
    dDdt = alpha * I
    return dSdt, dIdt, dRdt, dDdt

# Initial conditions vector
y0 = S0, I0, R0, D0
# Integrate the SIR equations over the time grid, t.
ret = odeint(deriv, y0, t, args=(N, beta, gamma, alpha))
S, I, R, D = ret.T
```

```python
# Plot the data on three separate curves for S(t), I(t) and R(t)
fig = plt.figure(facecolor='w',figsize=(20,10))
ax = fig.add_subplot(111, axisbelow=True)
ax.plot(t, S/N, 'b', alpha=0.5, lw=2, label='Susceptible')
ax.plot(t, I/N, 'y', alpha=0.5, lw=2, label='Infected')
ax.plot(t, R/N, 'g', alpha=0.5, lw=2, label='Recovered with immunity')
ax.plot(t, D/N, 'r', alpha=0.5, lw=2, label='Deceased')
ax.set_xlabel('Time /days')
ax.set_ylabel('Normalized Population/Cases number')
#ax.set_ylim(0,1.2)
ax.yaxis.set_tick_params(length=0)
ax.xaxis.set_tick_params(length=0)
ax.grid(b=True, which='major', c='w', lw=2, ls='-')
legend = ax.legend()
legend.get_frame().set_alpha(0.5)
for spine in ('top', 'right', 'bottom', 'left'):
    ax.spines[spine].set_visible(False)
plt.show()
```
