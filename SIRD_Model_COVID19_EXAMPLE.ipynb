```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Total population, N.
N = (1377492805 * 0.0002)
#N = 28000*600
print("N = ",N)
# Initial number of infected and recovered individuals, I0 and R0.
I0, R0, D0 = 1, 0, 0
# Everyone else, S0, is susceptible to infection initially.
S0 = N - I0 - R0 - D0
# Contact rate, beta, and mean recovery rate, gamma, (in 1/days).
beta, gamma,alpha = 0.277, 0.138, 0.0125
# A grid of time points (in days)
t = np.linspace(0, 200, 200)

# The SIR model differential equations.
def deriv(y, t, N, beta, gamma, alpha):
    S, I, R, D = y
    dSdt = -beta * S * I / N
    dIdt = beta * S * I / N - gamma * I - alpha * I
    dRdt = gamma * I
    dDdt = alpha * I
    return dSdt, dIdt, dRdt, dDdt

# Initial conditions vector
y0 = S0, I0, R0, D0
# Integrate the SIR equations over the time grid, t.
ret = odeint(deriv, y0, t, args=(N, beta, gamma, alpha))
S, I, R, D = ret.T
```

```python
# Plot the data on three separate curves for S(t), I(t) and R(t)
fig = plt.figure(facecolor='w',figsize=(20,10))
ax = fig.add_subplot(111, axisbelow=True)
ax.plot(t, S/N, 'b', alpha=0.5, lw=2, label='Susceptible')
ax.plot(t, I/N, 'y', alpha=0.5, lw=2, label='Infected')
ax.plot(t, R/N, 'g', alpha=0.5, lw=2, label='Recovered with immunity')
ax.plot(t, D/N, 'r', alpha=0.5, lw=2, label='Deceased')
ax.set_xlabel('Time /days')
ax.set_ylabel('Number (1000s)')
#ax.set_ylim(0,1.2)
ax.yaxis.set_tick_params(length=0)
ax.xaxis.set_tick_params(length=0)
ax.grid(b=True, which='major', c='w', lw=2, ls='-')
legend = ax.legend()
legend.get_frame().set_alpha(0.5)
for spine in ('top', 'right', 'bottom', 'left'):
    ax.spines[spine].set_visible(False)
plt.show()
```

```python

import numpy as np
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
from datetime import timedelta, date
plt.style.use('seaborn-pastel')


fig = plt.figure(figsize=(20,10))
ax = plt.axes(xlim=(0, 200), ylim=(-0.2, 1.2))
ax.set_title('SIRD model on COVID-19 Outbreak in India')
ax.set_xlabel('Time (in days)')
ax.set_ylabel('Number of cases')

line, = ax.plot([], [], 'b', alpha=0.5, lw=2, label='Susceptible')
line1, = ax.plot([], [], 'y', alpha=0.5, lw=2, label='Infected')
line2, = ax.plot([], [], 'g', alpha=0.5, lw=2, label='Recovered with immunity')
line3, = ax.plot([], [], 'r', alpha=0.5, lw=2, label='Deceased')

legend = ax.legend()
legend.get_frame().set_alpha(0.5)
t = np.linspace(0, 200, 200)
start_date = date(2020,1,30)
dates = [start_date+timedelta(days=n) for n in range(200)]
def init():
    line.set_data([], [])
    line1.set_data([], [])
    line2.set_data([], [])
    line3.set_data([], [])
    return line,line1,line2,line3
def animate(i):
    line.set_data(t[:i], S[:i]/N)
    line1.set_data(t[:i], I[:i]/N)
    line2.set_data(t[:i], R[:i]/N)
    line3.set_data(t[:i], D[:i]/N)
    return line,line1,line2,line3

anim = FuncAnimation(fig, animate, init_func=init,
                               frames=200, interval=20, blit=True)


anim.save('SIR_Model.mp4',fps=20)
```
